```python
#import pandas to read the file

import pandas as pd
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
# Transforming the nominal data to numeric data

dataset = pd.get_dummies(dataset, drop_first = True)
dataset
```

```python
# Getting the column names

dataset.columns
```

```python
# Splitting the input and the output data

independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
independent
```

```python
dependent = dataset[['Profit']]
dependent
```

```python
# Splitting the train and the test data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, train_size = 1/3, random_state= 0)
```

```python
# Preprocessing the input data using StandardScaler

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Preprocessing the output data using StandardScaler

scy=StandardScaler()
y_train = scy.fit_transform(y_train)
y_test = scy.transform(y_test)
```

```python
# Creating the model by importing the GridSearchCV and SVR

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = param_grid = {'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'gamma':['scale', 'auto'],
              'C':[10,100,1000,2000,3000]}
grid =GridSearchCV(SVR(), param_grid, refit = True, verbose = 3, n_jobs = -1)
grid.fit(x_train,y_train)
```

```python
# Printing the Best model parameters

re = grid.cv_results_
print("The Best Model Parameters are : {} ".format(grid.best_params_))
```

```python
# Creating the table view of the grid search

table = pd.DataFrame.from_dict(re)
table
```

```python
# Importing pickle to save the model and creating the filename

import pickle
filename = "final_model_svr_inout.sav"

pickle.dump(grid.best_estimator_, open(filename,'wb'))
```

```python
# preprocessing the input before loading to the model

prein = sc.transform([[1200,1300,4000,0,1]])
prein
```

```python
#Predicting the output using the model created

loaded_model = pickle.load(open("final_model_svr_inout.sav",'rb'))
result = loaded_model.predict(prein)
result
```

```python
# reverse processing the output before shown to the user

preout = scy.inverse_transform([result])
preout
```
