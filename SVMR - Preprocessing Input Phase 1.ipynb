```python
#import pandas to read the file

import pandas as pd
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
# Transforming the nominal data to numeric data

dataset = pd.get_dummies(dataset, drop_first = True)
dataset
```

```python
# Getting the column names

dataset.columns
```

```python
# Splitting the input and the output data

independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
independent
```

```python
dependent = dataset[['Profit']]
dependent
```

```python
# Splitting the train and the test data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
# Preprocessing the data using StandardScaler

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Creating the model by importing the GridSearchCV and SVR

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
#regressor = SVR()
#regressor.fit(x_train,y_train)
param_grid = {'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'gamma':['scale', 'auto'],
              'C':[10,100,1000,2000,3000]}
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3, n_jobs = -1)
grid.fit(x_train, y_train)
```

```python
# Printing the Best model parameters

re = grid.cv_results_
print("The Best model parameters are {} ".format(grid.best_params_))
```

```python
# Creating the table view of the grid search

table = pd.DataFrame.from_dict(re)
table
```

```python
# Importing pickle to save the model and creating the filename

import pickle
filename = "final_model_svr.sav"

pickle.dump(grid.best_estimator_,open(filename,'wb'))
```

```python
# preprocessing the input before loading to the model

preinput = sc.transform([[1300,1200,4000,0,1]])
preinput
```

```python
#Predicting the output using the model created

loaded_model = pickle.load(open("final_model_svr.sav",'rb'))
result = loaded_model.predict(preinput)
result
```
